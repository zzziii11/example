# Mission2. 가즈아!

```python
bitcoin = pd.read_csv('./BitCoin.csv').loc[:,['Date','Open']]

bitcoin
```

```python
bitcoin_sort = bitcoin.sort_values(by=['Date']).reset_index()

bitcoin_sort
```

```python
bitcoin2=bitcoin_sort.iloc[1130:1507,:]
bitcoin2
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=bitcoin2.loc[:,'Date']
y=bitcoin2.loc[:,'Open'].rolling(window=5).mean()
plt.xlabel('Date')
plt.ylabel('moving_average')


plt.title('5-MA BitCoin Graph')
plt.plot(x,y, color = 'y')

plt.show()
```

```python
eth = pd.read_csv('./ETH_day.csv').loc[:,['Date','Open']].sort_values(by=['Date']).reset_index().iloc[23:418,:]
eth
```

```python
x=bitcoin2.loc[:,'Date']
y=bitcoin2.loc[:,'Open'].rolling(window=5).mean()
x1=eth.loc[:,'Date']
y1=eth.loc[:,'Open'].rolling(window=5).mean()

plt.xlabel('Date')
plt.ylabel('moving_average')

plt.title('5-MA BitCoin & ETH Graph')
plt.plot(x,y, color = 'y')
plt.plot(x1,y1,color = 'k')

plt.show()
```
